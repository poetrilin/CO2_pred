```python
'''
@brief: visualization  
'''
import pandas as pd
import numpy as np
import os

os.getcwd()
```

```python
filepath = os.path.join(os.path.dirname(os.getcwd()),'dataset','train.csv')
train_df = pd.read_csv(filepath)
train_df.head()
```

```python
# 基本信息 columns 
train_df.info()  # 获取数据的基本信息，包括列名、数据类型和非空值数量等
train_df.describe()  # 获取数据的统计概览，包括计数、均值、标准差、最小值、最大值等

```

```python

import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("darkgrid")
sns.set_context("paper")
sns.set(font='SimHei',font_scale=1.5) 
plt.rcParams['axes.unicode_minus']=False
plt.rcParams['figure.figsize'] = (15, 5)
plt.rcParams['font.sans-serif']=['SimHei'] 


# 1. emmission和week_no的关系，可视化
# 先取同一longitude的数据,取第一个location_id
train_groupby_longitude = train_df.groupby('longitude')
first_group_key = list(train_groupby_longitude.groups.keys())[7]
group_1 = train_groupby_longitude.get_group(first_group_key)

sns.lineplot(x='week_no', y='emission', data=group_1)

plt.title('emission和week_no的关系')
plt.show()


```
